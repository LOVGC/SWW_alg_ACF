# Line Plots

```python
import plotly.graph_objects as go

df_stocks = px.data.stocks()
px.line(
    df_stocks, x="date", y="GOOG", labels={"x": "Date", "y": "Price"}
)  # the labels control the axes names

px.line(
    df_stocks,
    x="date",
    y=["GOOG", "AAPL"],
    labels={"x": "Date", "y": "Price"},
    title="Apple vs. Google",
)

```

```python
# can also create line plot using go

fig = go.Figure()  # create a figure
fig.add_trace(
    go.Scatter(x=df_stocks.date, y=df_stocks.AAPL, mode="lines", name="Apple")
)  # the name attribute is the legend

fig.add_trace(
    go.Scatter(x=df_stocks.date, y=df_stocks.AMZN, mode="lines + markers", name="AMZN")
)

fig.add_trace(
    go.Scatter(
        x=df_stocks.date,
        y=df_stocks.GOOG,
        mode="lines + markers",
        name="GOOG",
        line=dict(color="firebrick", width=2, dash="dashdot"),
    )  # line defines styles
)

# further style the figure
# fig.update_layout(
#     title="Stock Price Data 2018 - 2020", xaxis_title="Price", yaxis_title="Date"
# )

fig.update_layout(
    xaxis=dict(
        title = "time",
        showline=True,  # The showline axis property controls the visibility of the axis line
        showgrid=False,
        showticklabels=True,
        linecolor="rgb(204,204,204)",
        linewidth=2,
        ticks="outside",
        tickfont=dict(family="Arial", size=20, color="rgb(82,82,82)"),
    ),
    yaxis=dict(

        title = 'money',
        showgrid=False,
        zeroline = False,
        showline = True,
        showticklabels = False,
    ),
    autosize=False,
    margin = dict(autoexpand=False, l = 100, r =20, t = 110), # control the margin of the plot, l: left, r:right, t: top
    showlegend=False, 
    plot_bgcolor = 'White'
)
```
